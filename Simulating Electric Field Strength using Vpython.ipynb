```python
# GlowScript 2.6 VPython
from vpython import *

oofpez = 9e9 # The electric constant.
qe = 1.6e-19 # Fundamental charge.
sf = 3.0e-16 # Scale factor. Try changing if you can't see the fields.
source = sphere(pos=vector(-0.5e-8,0,0), radius = 1e-9,
                color=color.red) # This is the point charge.
q_source = qe # The point charge's charge value.
thetamax = 2*pi # The range of where to circle around the charge.
dtheta = pi/6 # The step with which to move around the charge.
R = 1.0e-8 # Radius of circle to move around the charge.

theta = 0 # Initiatlize theta.
while theta < thetamax: # Set up loop around charge.
#     rate(500)
    r_obs = R*vector(cos(theta),sin(theta),0) # Set observation point.
    r = r_obs - source.pos # Vector from source to observation point.
    rhat = r/mag(r) # Unit vector for direciton.
    E = (oofpez * q_source / mag(r)**2) * rhat # Electric field.
    arrow(pos=r_obs, color=color.orange, axis=sf*E,
          shaftwidth = source.radius) # Electric field arrow.
    theta = theta + dtheta # Increment theta.
```
